```python
# earlier we established that in order of medical charges, it was:
# non smoker < low BMI < high BMI < smoker
# we would like to see how many high bmi or smoker individuals can incur more than mean medical charges
# the mathematical notation should looks like this
```

```python
# data filter to set up the smoker & high cost
filter_smoker_high_cost = (df["smoker"] == 'yes') & (df["charges"] > 9382)
smoker_high_cost = df[filter_smoker_high_cost]["smoker"].count()

# high cost list
high_cost = df[df['charges'] > 9382]['charges'].count()

# result
p_smoker_high_cost = (smoker_high_cost/high_cost * 100).round(1)
print(p_smoker_high_cost)
```

```python
# data filter to set up the high BMI & high cost
filter_bmi_high_cost = (df["bmi"] > 24.9) & (df["charges"] > 9382)
bmi_high_cost = df[filter_bmi_high_cost]["smoker"].count()

# high cost list
high_cost = df[df['charges'] > 9382]['charges'].count()

# result
p_bmi_high_cost = (bmi_high_cost/high_cost * 100).round(1)
print(p_bmi_high_cost)
```

```python
# previously we know that mean of smoker medical charges is higher than mean of high BMI medical charges
# in this analysis, we conclude instead that: high BMI leads to greater number of individuals paying above mean medical cost
# compared to smokers
# my conclusion is: diseases generated by smoking is much more severe, but not as frequent as that coming from having high BMI
# also because high BMI have larger population compared to smokers (1096 : 274)
```
